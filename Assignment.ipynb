```python
#Problem 1: Estimating the U.S. Population
def population(year,last_digits_count=2):
   if(year>=2000):
     C=int(str(year)[last_digits_count:])
     D= (C-10) *3+ 310
     print("Population Result:" , D)
   else:
    print(-1)
population(2001)
population(2010)
population(2016)
population(1999)
```

```python
#Problem 2: Restaurant Tipping
def tip_amount(bill, good_service):
  if(bill<=25):
    tip=6.00
    print ("Tip value:", tip)
  elif(bill>25 and bool(good_service)==True):
    tip= bill* (27/100)
    print("Tip value:", tip)
  elif(bill>25 and bool(good_service)==False):
    tip= bill* (13/100)
    print("Tip value:", tip)
  else:
    print("NONE")

tip_amount(23.76, True)
tip_amount(23.37, False)
tip_amount(81.75, True)
tip_amount(63.59, True)
tip_amount(63.59, False)
```

```python
#Problem 3:Time Conversions
def getSeconds(hours,minutes,seconds):
  if(0<=int(hours)<=99 and 0<=int(minutes)<=59 and 0<=int(seconds)<=59):
    c= (int(hours)*3600) + (int(minutes)*60)+int(seconds)
    print("Return Value:", c)
  else:
    print("Declared right value")
getSeconds("11", "11", "14")
getSeconds("00", "01", "02") 
getSeconds("05", "00", "30")
getSeconds("00", "00", "00") 
getSeconds("02", "18", "49")
```

```python
#Problem 4: Temperature Conversion
def temp_converter(temperature,deg):
  if(deg=="C"):
    C_temp= (float(temperature)-32) * 5  / 9 
    print(C_temp, "degree Celsius")
  elif(deg=="F"):
    F_temp=(float(temperature)* 1.8) + 32
    print(F_temp, "degree Farenheit")
  else:
    print("Error: Select F or C ")
temp_converter(100, "C")
temp_converter(0, "F")
temp_converter(212, "C")
temp_converter(32, "C")
temp_converter(50, "F")
temp_converter(100,"A")
```

```python
#Problem 5: The Depressingly Large and Slow Universe
def how_long(distance,fraction,scale):
  c= 186000
  speed= c*fraction
  time_in_sec= distance/speed
  if (scale == 'M'):
    Minutes=time_in_sec / 60
    print(Minutes)
  elif (scale == 'H'):
    Hour= time_in_sec / 3600
    print(Hour)
  elif (scale == 'D'):
    Day= time_in_sec / (3600*24)
    print(Day)
  elif (scale == 'Y'):
    Year= time_in_sec / (3600*24*365)
    print(Year)
  else:
    print("Select correct scale value for calculation")

how_long(238900, 0.01, 'M')
how_long(45594000, 0.01, 'H')
how_long(93000000, 1.0, 'M')
how_long(9000000000, 0.01, 'D')
how_long(25000000000000, 0.01, 'Y')
```
